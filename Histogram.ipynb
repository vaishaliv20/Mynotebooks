# HISTOGRAM- Shows how many times values are repeating/appearing and is useful in understanding the distribution of numeric variables.

For numeric columns- (Multicolumns analysis)-  df.hist()

For numeric columns- (Univariate analysis)-  plt.hist(); sns.displot()

For numeric+categorical column- (Bivariate analysis)- (X-axis represents numeric column)- sns.displot()

For categorical+numeric columns (Multivariate analysis)-(X-axis represents numeric column)-  sns.displot()

For categorical+numeric column- (Bivariate analysis)- (X-axis represents categorical/numeric column)- sns.histplot(), sns.FacetGrid()

For numeric columns- (Univariate analysis)-  sns.displot(); plt.hist()

```python
### For 1 numeric column(suitable for smaller dataset)-X-axis-numeric column's values; Y-axis-count(frequencies) of those values
# importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
# loading exercise dataset provided by seaborn
exercise = sns.load_dataset("exercise")
# making plot using displot()
sns.displot(data=exercise, x='pulse', kind='hist')  #with keyword 'hist'
#OR
sns.displot(data=exercise, x='pulse', bins=13)   #without keyword
# visualizing the plot using matplotlib.pyplot show() function
plt.show()
```

For numeric+categorical column- (Bivariate analysis)- (X-axis represents numeric column)- sns.displot(); sns.FacetGrid()

```python
### 2 columns--1 numeric column on X-axis; 1 cat. column as 'Col'; count on Y-axis
# importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
# loading exercise dataset provided by seaborn
exercise = sns.load_dataset("exercise")
sns.displot(data=exercise, x='pulse', col='kind', kind='hist')
# visualizing the plot using matplotlib.pyplot show() function
plt.show()
#OR
#Subplots based on different categories of 'col'
#1 cont. and 1 cat. var.
g = sns.FacetGrid(exercise, col="kind") #count on x-axis
g.map(sns.histplot, "pulse")  #pulse (cont var) on y-axis
```

For categorical+numeric column- (Bivariate analysis)- (X-axis represents categorical column)- sns.histplot()

```python
### 2 columns--X-axis represents cat. column; Y-axis represents numeric column
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
# loading dataset
data = sns.load_dataset("iris")
sns.histplot(x='species', y='sepal_width', data=data)
plt.show()
```

For categorical+numeric columns (Multivariate analysis)-(X-axis represents numeric column)-  sns.displot()

```python
#### Subplots-
### 3 columns (2 cat.+1 num.)--1 numeric column on X-axis; 1 cat. column as 'Col'; other cat. column as 'Row'; count on Y-axis
# importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
# loading exercise dataset provided by seaborn
exercise = sns.load_dataset("exercise")
sns.displot(data=exercise, x='pulse', col='kind', row='time', kind='hist')
# visualizing the plot using matplotlib.pyplot show() function
plt.show()
```

```python
### 3 columns (2 cat.+1 num.)--1 numeric column on X-axis; 1 cat. column as 'Col'; other cat. column as 'Hue'; count on Y-axis
# importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
# loading exercise dataset provided by seaborn
exercise = sns.load_dataset("exercise")
sns.displot(data=exercise, x='pulse', col='time', hue='kind', kind='hist', multiple="dodge")  #use of multiple to avoid overlapping in bars
# visualizing the plot using matplotlib.pyplot show() function
plt.show()
```

# COUNT PLOT- Gives count of categorical data in graph form

Show value counts for a single categorical variable- sns.counplot(x), sns.catplot(kind='count')

Show value counts for two categorical variables (using hue parameter)- sns.countplot(hue=''), sns.catplot(kind='count')

Show value counts for a single categorical variable- sns.counplot(x), sns.catplot(kind='count')

```python
# importing the required library
import seaborn as sns
import matplotlib.pyplot as plt
# read a tips.csv file from seaborn library
df = sns.load_dataset('tips')
# count plot on single categorical variable
sns.countplot(x ='sex', data = df)    #default count on y-axis
# Show the plot
plt.show()
```

```python
import seaborn as sns
sns.set_theme(style="ticks")
exercise = sns.load_dataset("exercise")
g = sns.catplot(x="time", kind="count", data=exercise)
```

Show value counts for two categorical variables (using hue parameter)- sns.countplot(hue=''), sns.catplot(kind='count')

```python
#2 categorical var.- one on x-axis; other as 'bars'/'hue'
# importing the required library
import seaborn as sns
import matplotlib.pyplot as plt
# read a tips.csv file from seaborn library
df = sns.load_dataset('tips')
# count plot on two categorical variable
sns.countplot(x ='sex', hue = "smoker", data = df)  #2 cat. var.-sex and smoker
# Show the plot
plt.show()
```

```python
#Plotting horizontally
# importing the required library
import seaborn as sns
import matplotlib.pyplot as plt
# read a tips.csv file from seaborn library
df = sns.load_dataset('tips')
# count plot along y axis
sns.countplot(y ='sex', hue = "smoker", data = df)
# Show the plot
plt.show()
```

```python
#2 cat. var.- one on x-axis; other as subplot column; count on y-axis
#condition specified for data in SQL form
titanic = sns.load_dataset("titanic")
g = sns.catplot(x="alive", col="deck", col_wrap=4, data=titanic[titanic.deck.notnull()], kind="count", height=2.5, aspect=.8) 
```
