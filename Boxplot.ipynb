# BOX PLOT

For 1 numeric column (Univariate analysis)- sns.boxplot(); sns.catplot(kind='box')

For categorical+numeric column (Bivariate analysis)- df.boxplot(by=''); sns.boxplot();  sns.catplot(kind='box')

For numeric+categorical columns (Multivariate analysis)- sns.boxplot(x,y,hue), sns.catplot(kind='box')

For numeric columns (Multicolumn analysis)- sns.boxplot();  sns.catplot(kind='box')

For 1 numeric column (Univariate analysis)- sns.boxplot()

```python
#Horizontal boxplot with numeric column on X-axis
import seaborn as sns
sns.set(style='whitegrid')
tip = sns.load_dataset("tips")
sns.boxplot(x =tip['total_bill'])
```

For categorical+numeric column (Bivariate analysis)- df.boxplot(by=''); sns.boxplot(), sns.catplot(kind='box')

```python
#Numeric column on y-axis; cat. column on x-axis
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
# load the dataset
df = pd.read_csv("tips.csv")
#Boxplot of days with respect total_bill
df.boxplot(by ='day', column =['total_bill'], grid = False)
```

```python
#Cat. column on X-axis; Numeric column (discrete) on Y-axis
import seaborn as sns
sns.set(style='whitegrid')
tip = sns.load_dataset("tips")
sns.boxplot(x="day", y="tip", data=tip)
```

```python
#Cat. column on X-axis; Numeric column on Y-axis
tips = sns.load_dataset('tips')
sns.catplot(x='day', y='total_bill', data=tips, kind='box');
```

```python
#Horizontal boxplot
# Numeric column on X-axis; Cat. column on Y-axis
import seaborn as sns
sns.set(style='whitegrid')
tip = sns.load_dataset("tips")
sns.boxplot(x="tip", y="day", data=tip)
```

For numeric+categorical columns (Multivariate analysis)- sns.boxplot(x,y,hue), sns.catplot(x,y,hue, kind='box')

```python
# 3 columns- 1 numeric on Y-axis, 1 cat. on X-axis and other cat. as 'hue'
# importing the required module
import seaborn
# use to set style of background of plot
seaborn.set(style="whitegrid")
# loading data-set
tip = seaborn.load_dataset("tips")
seaborn.boxplot(x="day", y="total_bill",hue="smoker",data=tip, palette="Set2",dodge=True)
```

```python
# 3 columns- 1 numeric on Y-axis, 1 cat. on X-axis and other cat. as 'hue'
# importing the required module
import seaborn
# use to set style of background of plot
seaborn.set(style="whitegrid")
# loading data-set
tip = seaborn.load_dataset("tips")
seaborn.catplot(x="day", y="total_bill", kind='box', hue="smoker",data=tip, palette="Set1")
```

For numeric columns (Multicolumn analysis)- sns.boxplot();  sns.catplot(kind='box')

```python
#Horizontal boxplot with multiple numeric column names on Y-axis and corresponding values on X-axis
# importing the required module
import seaborn
# use to set style of background of plot
seaborn.set(style="whitegrid")
# loading data-set
tip = seaborn.load_dataset("tips")
seaborn.boxplot(data = tip,orient="h")  #orient represents orientation
```

# VIOLIN PLOT

For 1 numeric column (Univariate analysis)- sns.violinplot()

For categorical+numeric column (Bivariate analysis)- sns.violinplot();  sns.catplot(kind='violin')

For numeric+categorical columns (Multivariate analysis)- sns.violinplot(x,y,hue/row/col), sns.catplot(x,y,hue/col/row)

For 1 numeric column (Univariate analysis)- sns.violinplot()

```python
import seaborn
# use to set style of background of plot
seaborn.set(style="whitegrid")
# loading data-set
tips = seaborn.load_dataset("tips")
seaborn.violinplot(x=tips["total_bill"])
```

For categorical+numeric column (Bivariate analysis)- sns.violinplot();  sns.catplot(kind='violin')

```python
import seaborn
seaborn.set(style = 'whitegrid') 
tip = seaborn.load_dataset('tips')
#Both the below codes give same graph
seaborn.violinplot(x ='day', y ='tip', data = tip)
#OR
seaborn.catplot(x ='day', y ='tip', data = tip, kind='violin')
```

```python
#Horizontal violinplot by interchanging x- and y-axis columns
# importing the required module
import seaborn
# use to set style of background of plot
seaborn.set(style="whitegrid")
# loading data-set
tips = seaborn.load_dataset("tips")
seaborn.violinplot(x="tip", y="day", data=tip)  #same code can be used with sns.catplot
```

For numeric+categorical columns (Multivariate analysis)- sns.violinplot(x,y,hue/row/col), sns.catplot(x,y,hue/col/row)

```python
# 3 columns- 1 numeric on Y-axis, 1 cat. on X-axis and other cat. as 'hue'
# importing the required module
import seaborn
# use to set style of background of plot
seaborn.set(style="whitegrid")
# loading data-set
tips = seaborn.load_dataset("tips")
seaborn.violinplot(x="day", y="total_bill", hue="time", data=tips)
```

```python
#Using catplot
# 3 columns- 1 numeric on Y-axis, 1 cat. on X-axis and other cat. as 'hue'
import seaborn as sns
exercise = sns.load_dataset("exercise")
g = sns.catplot(x="time", y="pulse", hue="kind", data=exercise, kind="violin")
```

```python
#Horizontal violinplot
#4 columns-1 numeric on X-axis, 1 cat. on Y-axis, 1 cat. as hue and 1 cat. as row
titanic = sns.load_dataset("titanic")
g = sns.catplot(x="age", y="embark_town",hue="sex", row="class",data=titanic[titanic.embark_town.notnull()], orient="h", height=2, aspect=3, palette="Set3",
                kind="violin", dodge=True, cut=0, bw=.2)
```

# BOXENPLOT

For 1 numeric column (Univariate analysis)- sns.boxenplot()

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
# loading dataset
data = sns.load_dataset("tips")
sns.boxenplot(x="total_bill", data=data)
```

For categorical+numeric column (Bivariate analysis)- sns.boxenplot();  sns.catplot(kind='boxen')

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
# loading dataset
data = sns.load_dataset("tips")
# plot the boxenplot
sns.boxenplot(x = "day", y = "total_bill", data = data)
plt.show()
#OR
sns.catplot(x = "day", y = "total_bill", data = data, kind='boxen')  #both code will give same graph
plt.show()
```

```python
#Horizontal boxenplot
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
# loading dataset
data = sns.load_dataset("tips")
# plot the boxenplot orient to horizontal
sns.boxenplot(x = "total_bill", y = "size", data = data, orient ="h")
plt.show()
```

For numeric+categorical columns (Multivariate analysis)- sns.boxenplot(x,y,hue/row/col), sns.catplot(x,y,hue/row/col, kind='boxen')

```python
# 3 columns- 1 numeric on Y-axis, 1 cat. on X-axis and other cat. as 'hue'
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
# loading dataset
data = sns.load_dataset("tips")
# plot the boxenplot hue by sex width of 0.8
sns.boxenplot(x ="day", y = "total_bill", hue = "sex", data = data, width = 0.8)  #catplot can also be used
plt.show()
```
