Scatter plot with regression line(by default)

```python
# importing the required library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
# scatter plot with regression line(by default)
sns.lmplot(x ='total_bill', y ='tip', data = tips)
# Show the plot
plt.show()
```

Scatter plot without regression line

```python
# importing the required library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.lmplot(x ='total_bill', y ='tip', fit_reg = False, data = tips)
# Show the plot
plt.show()
```

```python
# importing the required library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# scatter plot using hue attribute for colouring out points according to the sex
sns.lmplot(x ='total_bill', y ='tip', fit_reg = False, hue = 'sex', data = tips)
# Show the plot
plt.show()
```

For 2 numeric+2 or more categorical columns (Multivariate analysis)- sns.FacetGrid(); sns.relplot(x,y,hue,col/row)

```python
#4 columns- 2 cont. and 2 cat. var.
#2 cont. var. on x and y axes and cat. as hue and col
import seaborn as sns
sns.set(style ="ticks")
tips = seaborn.load_dataset("tips")
g = sns.FacetGrid(tips, col="sex", hue="smoker")
g.map(sns.scatterplot, "total_bill", "tip", alpha=.7)
g.add_legend()
```

```python
#5 columns- 2 numeric on X- and Y-axis; 1 cat. as hue; 1 cat. as row; 1 cat. as col
# importing the library
import seaborn as sns
# selecting style
sns.set(style ="ticks")
# reading the dataset
tips = sns.load_dataset('tips')
sns.relplot(x="total_bill",y="tip",hue="day",col="time",row="sex",data=tips)
```

# STRIP PLOT

For 1 numeric+2 or more categorical columns (Multivariate analysis)- sns.catplot()

For numeric+categorical column- (Bivariate analysis)- sns.stripplot()

For numeric+categorical column- (Bivariate analysis)- sns.stripplot()

```python
#Cat. column on X-axis; numeric column on Y-axis
import seaborn
import matplotlib.pyplot as plt
seaborn.set(style = 'whitegrid')  
tip = seaborn.load_dataset("tips")  
seaborn.stripplot(x="day", y="total_bill", data=tip)
plt.show()
```

For 1 numeric+2 or more categorical columns (Multivariate analysis)- sns.catplot(); sns.stripplot(x,y,hue)

```python
#3 columns- 1 numeric on Y-axis; 1 cat. on X-axis; 1 cat. as hue
#Using sns.stripplot
# importing the required module
import seaborn
import matplotlib.pyplot as plt
# use to set style of background of plot
seaborn.set(style = 'whitegrid')
# loading data-set
tips = seaborn.load_dataset("tips")
seaborn.stripplot(x="sex", y="total_bill", hue="day", data=tips)
```

```python
#Clear distinction in categories' data points with 'dodge' parameter
#3 columns- 1 numeric on Y-axis; 1 cat. on X-axis; 1 cat. as hue
seaborn.stripplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set2", dodge=True)
```

```python
#3 columns- 1 numeric on Y-axis; 1 cat. on X-axis; 1 cat. as hue
#Using sns.catplot
import seaborn as sns
exercise = sns.load_dataset("exercise")
g = sns.catplot(x="time", y="pulse", hue="kind", data=exercise)
```

```python
### Subplots-
#4 columns- 1 numeric column on Y-axis; 1 cat. on X-axis; 1 cat. as hue; 1 cat. as col
import seaborn as sns
exercise = sns.load_dataset("exercise")
g = sns.catplot(x="time", y="pulse", hue="kind", col="diet", data=exercise)
```

# SWARM PLOT

For numeric+categorical column- (Bivariate analysis)-  sns.swarmplot();  sns.catplot(kind='swarm')

For numeric+2 or more categorical columns- (Multivariate analysis)-  sns.swarmplot(x,y, hue, col/row);  sns.catplot(kind='swarm', col/row)

For numeric+categorical column- (Bivariate analysis)-  sns.swarmplot();  sns.catplot(kind='swarm')

```python
#Cat. column on X-axis; numeric column on Y-axis
import seaborn
import matplotlib.pyplot as plt
seaborn.set(style = 'whitegrid')  
tip = seaborn.load_dataset("tips") 
#Below codes will give same graph
seaborn.swarmplot(x="day", y="total_bill", data=tip)
plt.show()
#OR
seaborn.catplot(x="day", y="total_bill", data=tip, kind='swarm')  #catplot with 'kind' type as 'swarm'
plt.show()
```

```python
#Horizontal swarmplot
# importing the required module
import seaborn
# use to set style of background of plot
seaborn.set(style="whitegrid")
# loading data-set
tips = seaborn.load_dataset("tips")
seaborn.swarmplot(x="total_bill", y="day", data=tips, size=6.5)
```

For numeric+2 or more categorical columns- (Multivariate analysis)-  sns.swarmplot(x,y, hue, col/row);  sns.catplot(kind='swarm', col/row)

```python
# importing the required module
import seaborn
# use to set style of background of plot
seaborn.set(style="whitegrid")
# loading data-set
tips = seaborn.load_dataset("tips")
seaborn.swarmplot(x="day", y="total_bill", hue="time", data=tips)  #can also use catplot with 'swarm' as 'kind'
```

```python
# importing the required module
import seaborn
# use to set style of background of plot
seaborn.set(style="whitegrid")
# loading data-set
tips = seaborn.load_dataset("tips")
seaborn.swarmplot(x="day", y="total_bill", hue="smoker",
                  data=tips, palette="Set2", dodge=True)  #use of 'dodge' for points distinctiveness
```
