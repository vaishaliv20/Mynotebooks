# BAR PLOT- Good to compare quantities corresponding to different groups

A bar plot is a plot that presents categorical data with
rectangular bars with lengths proportional to the values that they
represent. A bar plot shows comparisons among discrete categories. One
axis of the plot shows the specific categories being compared, and the
other axis represents a measured value.

For numeric columns (Multi-columns analysis)

For categorical+numeric columns (Multivariate analysis)- (y-axis represents numeric column)

For numeric/categorical columns (Bivariate analysis)- (y-axis represents numeric column)

# Bar plot codes-

For numeric columns (Multi-columns analysis)-  df.plot.bar()

```python
### X-axis--index(default)/'column names'(if specified as 'x'); Y-axis--respective numeric values; bars represent column names

###For very few columned-dataframe (approx 2 to 8 columns)
df.plot.bar()     #no parameters passed(optional); index on x-axis since x is not specified.
df.plot.barh()   #Horizontal bar plot
df.plot.bar(stacked=True)   #stacked bar plot
#[["Not recommended for numeric columns" 
df.plot.bar(x='a', rot=30) ]] #values of 'a' column as scale on x-axis (instead of default 'index') and plots other columns as bars and rotates the x-labels

df.iloc[2].plot(kind="bar")  #values on y-axis represent values for row number 2 for each column; column names on x-axis

#Subplots (of all columns)-
df.plot.bar(subplots=True, color={'a': 'green', 'b': 'red', 'c': 'blue'})  #separate bar plot for each column with index on x-axis along with defining color for each column
#Subplots (of specific columns)-
df.iloc[:,0:2].plot.bar(subplots=True)   

###For larger dataset to check for specific column(s)
df.iloc[5].plot.bar()    #For specific columns based on index
df.iloc[5].plot(kind="bar")   #'kind' keyword used with matplotlib's plot() method
df.iloc[:,0:5].plot.bar(subplots=True)   ###Subplots of specific columns
```

```python
import matplotlib.pyplot as plt # Import Required Package
Days = [1,2,3,4,5,6,7,8]
Speed = [60,62,61,58,56,57,46,63]
plt.plot(Days, Speed, color = 'blue') # For Create a Line Plot
plt.xlabel('Days')
plt.ylabel('Car Speed')
plt.title('Car Speed Measurement')
plt.show()
```

```python
Days = [1,2,3,4,5,6,7,8]

Speed = [60,62,61,58,56,57,46,63]
plt.plot(Days, Speed, color = 'blue', marker = 'o', ms = 20, mec = 'red', mfc = 'green', linestyle = 'dotted', linewidth = 3.5)
plt.xlabel('Days') # ms for marker size, mfc for marker face color, mec for border color

plt.ylabel('Car Speed')
plt.title('Car Speed Measurement')
plt.show()
```

```python
days = [1,2,3,4,5,6,7,8]
max_speed = [80,91,92,88,77,79,76,75]
min_speed = [42,43,40,42,33,36,34,35]
avg_speed = [46,58,57,56,40,42,41,36]
plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
plt.plot(Days, max_speed, color = 'red', marker = '*', ms = 10, mec = 'red', mfc = 'orange', label = 'Maximum Speed')
plt.xlabel('Days')

plt.ylabel('Maximum Speed')
plt.title('Car Speed Measurement')
plt.legend()
plt.grid()
plt.show()
plt.figure(figsize=(8,6))
plt.subplot(2,1,2)
plt.plot(Days, min_speed, color = 'green', marker = 'o', ms = 10, mec = 'black', mfc = 'blue', label = 'Minimum Speed')
plt.xlabel('Days')
plt.ylabel('Minimum Speed')
plt.title('Car Speed Measurement')
plt.legend()
plt.grid()
plt.show()
plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
plt.plot(Days, avg_speed, color = 'yellow', marker = '^', ms = 10, mec = 'red', mfc = 'orange', label = 'Average Speed')
plt.xlabel('Days')
plt.ylabel('Maximum Speed')
plt.title('Car Speed Measurement')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
Marks = np.array([61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64])
below_avg = Marks[np.logical_and(Marks >= 40, Marks < 60)]
avg_marks = Marks[np.logical_and(Marks >= 60, Marks < 80)]
exe_marks = Marks[np.logical_and(Marks >= 80, Marks <= 100)]
plt.hist(Marks)
plt.show()
print('Below Average Students Are : ',below_avg.size)
print('Average Students Are : ',avg_marks.size)
print('Execelent Students Are : ',exe_marks.size)
```
