# DISTRIBUTION PLOTS

# Pairplot- It represents pairwise relation across the entire dataframe and supports an additional argument called hue for categorical separation.

For pairwise relationship between numeric columns- sns.pairplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt 
# to ignore the warnings 
from warnings import filterwarnings
# load the dataset
df = sns.load_dataset('tips')
sns.pairplot(df)
```

For pairwise relationship between numeric columns with one categorical column- sns.pairplot(hue='')

```python
#Numeric columns (pair-wise) relationship with one categorical column as 'hue'
#'hue' sets up the categorical separation between the entries of the dataset.
#'palette' is used for designing the plots.
# import the necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt 
# to ignore the warnings 
from warnings import filterwarnings
# load the dataset
df = sns.load_dataset('tips')
sns.pairplot(df, hue ="sex", palette ='coolwarm')
```

# KDE plots

For numeric+categorical column (Bivariate analysis)- sns.displot(kind='kde', col='')

For numeric+categorical columns (Multivariate analysis)- sns.displot(kind='kde', col='', row='')

For numeric+categorical column values (Bivariate analysis- with SQL condition)- sns.kdeplot()

For numeric column (Univariate analysis)- sns.displot(kind='kde'), sns.kdeplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
# loading exercise dataset provided by seaborn
exercise = sns.load_dataset("exercise")
# making plot using displot()
sns.displot(data=exercise, x='pulse', kind='kde')
```

```python
### With SQL condition on cat. column values to select for a numeric column
# importing the required libraries
from sklearn import datasets
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from warnings import filterwarnings
# Setting up the Data Frame
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=['Sepal_Length','Sepal_Width', 'Patal_Length', 'Petal_Width'])
iris_df['Target'] = iris.target
iris_df['Target'].replace([0], 'Iris_Setosa', inplace=True)
iris_df['Target'].replace([1], 'Iris_Vercicolor', inplace=True)
iris_df['Target'].replace([2], 'Iris_Virginica', inplace=True)
# Plotting the KDE Plot
sns.kdeplot(iris_df.loc[(iris_df['Target']=='Iris_Virginica'),'Sepal_Length'], color='b', shade=True, Label='Iris_Virginica')  #SQL condition
# Setting the X and Y Label
plt.xlabel('Sepal Length')
plt.ylabel('Probability Density')
```

For numeric+categorical column values (Bivariate analysis- with SQL condition)- sns.kdeplot()

```python
#1 numeric column but for specific values of cat. column in the same graph; using SQL for conditioning to select rows
# Plotting the KDE Plot
#Condition on cat. column for desirable data related to its values
sns.kdeplot(iris_df.loc[(iris_df['Target']=='Iris_Setosa'),'Sepal_Length'], color='r', shade=True, Label='Iris_Setosa')
sns.kdeplot(iris_df.loc[(iris_df['Target']=='Iris_Virginica'), 'Sepal_Length'], color='b', shade=True, Label='Iris_Virginica')
plt.xlabel('Sepal Length')
plt.ylabel('Probability Density')
```

For numeric+categorical column (Bivariate analysis)- sns.displot(kind='kde', col='')

```python
#### Subplots-
### 2 columns--1 numeric column on X-axis; 1 cat. column as 'Col'; density on Y-axis
# importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
# loading exercise dataset provided by seaborn
exercise = sns.load_dataset("exercise")
sns.displot(data=exercise, x='pulse', col='kind', kind='kde')
plt.show()
```

For numeric+categorical columns (Multivariate analysis)- sns.displot(kind='kde', col='', row='')

```python
#### Subplots-
### 3 columns (2 cat.+1 num.)--1 numeric column on X-axis; 1 cat. column as 'Col'; other cat. column as 'Row'; density on Y-axis
# importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
# loading exercise dataset provided by seaborn
exercise = sns.load_dataset("exercise")
sns.displot(data=exercise, x='pulse', col='kind', row='time', kind='kde')
```
